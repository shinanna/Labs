```python
import pandas as pd
import numpy as np
```

```python
df_ori = pd.read_excel(r"C:\Users\Shinemet\Ironhack\Labs\Lab32\example_data_cleaning.xlsx")
```

```python
# Data preparation plan: 
## create and work on a copy of my dataset
## look at my dataset statistics
## deal with missing values, identify outliers, harmonize 'Profession' values, drop duplicates, encode categorical values
```

```python
df = df_ori.copy()
df
```

```python
desc = pd.DataFrame(df.describe(), dtype=np.int64)
desc
```

```python
# no variance on column Department ==> drop
# outliers identified at row 19 (max Amount 2399090) ==> drop
# other outliers identified at row 1 (BirthYear 2013, min Amount 1230) and row 17 (max BirthYear 2017): 
# let's assume the column is not labeled properly (1st year of worklife instead of BirthYear e.g.) ==> keep 

df.drop(19, inplace=True)
df.drop(['Department'], axis=1, inplace=True)
df
```

```python
# compute new statistics
desc = pd.DataFrame(df.describe(), dtype=np.int64)
desc
```

```python
df.isna().sum()
```

```python
# replace BirthYear NaN values by the mean
by_mean = df['BirthYear'].mean()
df['BirthYear'].fillna(by_mean, inplace=True)
df
```

```python
# replace missing values with relevant data found on same ClientID
df['BirthYear'][7]=df['BirthYear'][6]
df['BirthYear'][8]=df['BirthYear'][9]
df['Profession'][28]=df['Profession'][27]
df['Profession'][29]=df['Profession'][27]
df
```

```python
# compute new statistics
desc = pd.DataFrame(df.describe(), dtype=np.int64)
desc
```

```python
# replace irrelevant values : typos, cases
df['Profession'] = df['Profession'].replace(['student','etudient'],'Student')
df['Profession'] = df['Profession'].replace(['manager'],'Manager')
df['Profession'] = df['Profession'].replace(['bdm'],'BDM')
df['Profession'] = df['Profession'].replace(['hr'], 'HR')
df
```

```python
# check any duplicates
df.duplicated().any()
```

```python
# encode categorical data found in 2 columns: Profession and Risk
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

df['Profession']= label.fit_transform(df['Profession'])
df['Risk']= label.fit_transform(df['Risk'])
```
