#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
# retrieving only image links relating to the page
r = requests.get(url)
r.status_code
html = r.content
soup = BeautifulSoup(html, 'lxml')
[link.get('href') for link in soup.find_all(class_='image')]
````

````python
# retrieving all image links including thumbnails etc.
[upload.get('src') for upload in soup.find_all('img')]
````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
r = requests.get(url)
r.status_code
html = r.content
soup = BeautifulSoup(html, 'lxml')
lst_changes = [title.text.strip(' \n/') for title in soup.find_all(class_='usctitlechanged')]

print('There has been %d change(s) since last release point:' % len(lst_changes))
print(lst_changes)
````

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````
