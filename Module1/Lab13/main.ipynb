```python
r = requests.get('https://api.github.com/search/code?q=.scavengerhunt in:path+repo:ironhack-datalabs/scavenger');
search = r.json()
search
```

```python
files = search['items'];
files = sorted(files, key=lambda file: file['name']);
print(files)
```

```python
contents_list = []
headers = {'Accept':'application/vnd.github.v3.raw', };
headers
```

```python
for file in files:
    content_path = ('https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{}').format(file['path']);
    r = requests.get(content_path, headers=headers);
    contents_list.append(r.text);
```
