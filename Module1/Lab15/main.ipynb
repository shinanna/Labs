#### Question 1: Find the questions and answers of last month.

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
nov_questions = SITE.fetch('questions', fromdate=1635724800, todate=1638230400)
nov_questions
```

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
nov_answers = SITE.fetch('answers', fromdate=1635724800, todate=1638230400)
nov_answers
```
