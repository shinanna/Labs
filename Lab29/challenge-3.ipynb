## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: group by Customer IDs to calculate unique customers aggregated amount_spent
orders_as = orders.groupby('CustomerID').agg({'amount_spent':'sum'})
orders_as
```

```python
# Sub Problem 2: group Customer IDs by amount_spent quantiles
orders_qt = orders_as.quantile(q=[.75,.95], axis=0)
orders_qt
```

```python
# Find min and max amount_spent to define cutoffs
orders_as.describe()
```

```python
# Extract customers between 75th-95th and in 95th percentile and label them in the df
customers = ['Preferred', 'VIP']
cutoffs = [1661.640, 5840.182, 280206.020000]
orders_as['Customer_type'] = pd.cut(orders_as['amount_spent'], cutoffs, labels=customers)
orders_as
```

# Q2: How to identify which country has the most VIP Customers?

```python
# Merge customer type df with initial country df
c_vip = orders_as.merge(orders[['CustomerID','Country']], on='CustomerID', how='outer')
c_vip
```

```python
# Filter VIP customers
filtered = c_vip[(c_vip['Customer_type']=='VIP')]
filtered
```

```python
# Sort countries by customers count: biggest VIP country is the UK
df = filtered.groupby(['Country']).agg({'CustomerID': 'count'})
df.sort_values(by='CustomerID', ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# Filter VIP + Preferred customers and sort the df by Country: the UK
filtered = c_vip[(c_vip['Customer_type']=='VIP')|(c_vip['Customer_type']=='Preferred')]

df = filtered.groupby(['Country']).agg({'CustomerID': 'count'})
df.sort_values(by='CustomerID', ascending=False)
```
