#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y = x**2
z = np.exp(x)
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(12,6))

ax1.plot(x,y, 'r')
ax1.set_title('Quadratic')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_xscale('linear')
ax1.set_yscale('linear')

ax2.plot(x,z, 'g')
ax2.set_title('Exponential')
ax2.set_xlabel('x axis')
ax2.set_ylabel('y axis')
ax2.set_xscale('log')
ax2.set_yscale('log')

plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x,y, color='red', label='Quadratic')

ax.set_title('Quadratic and Exponential')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_xscale('linear')

ax2 = ax.twinx()
ax2.plot(x,z, color='green', label='Exponential')
ax2.set_yscale('log')

lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc=0)

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
sbm = fitbit[['Steps', 'Months_encoded']].groupby(['Months_encoded']).agg('median')
sbm.reset_index(inplace=True)

# MEDIAN STEPS BY WORK_OR_WEEKEND
swk = fitbit[['Steps', 'Work_or_Weekend']].groupby(['Work_or_Weekend']).agg('median')
swk.reset_index(inplace=True)
```

Plotting the 2 variables don't show any obvious correlation to my sense. 
Let's plot a correlation matrix of the dataset's numerical variables. 

```python
fig = plt.figure(figsize=(8, 8))
plt.matshow(titanic.corr(), fignum=fig.number)
plt.xticks(range(titanic.select_dtypes(['number']).shape[1]), titanic.select_dtypes(['number']).columns, fontsize=10, rotation=45)
plt.yticks(range(titanic.select_dtypes(['number']).shape[1]), titanic.select_dtypes(['number']).columns, fontsize=10)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=10)
plt.title('Correlation Matrix', y=1.1, verticalalignment='bottom', fontsize=12)

plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
ax = sns.heatmap(
    titanic.corr(), 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)

plt.show()
```

#### What are the most correlated features?

Pclass and Fare, which makes sense since classes are defined by fares.
